### Data Processing

```python
#Checking for missing values
df.isna().sum()
```

```python
#Correcting the eeror values in the df['cutoff_timestamp']
df['cutoff_timestamp'] = df['cutoff_timestamp'].apply(lambda x: x[:19])
```

```python
#Converting timestamp to datetime
time_columns = ['trip_creation_time', 'od_start_time', 'od_end_time', 'cutoff_timestamp']
for col in time_columns:
    df[col] = pd.to_datetime(df[col])
```

### Feature engineering

```python
#Creating the segment_key column from trip_uuid, source_center, and destination_center
df['segment_key'] = df['trip_uuid'] + df['source_center'] + df['destination_center']
```

```python
#Creating new columns segment_actual_time_sum, segment_osrm_distance_sum, segment_osrm_time_sum
df['segment_actual_time_sum'] = df.groupby(by='segment_key')['segment_actual_time'].cumsum()
df['segment_osrm_distance_sum'] = df.groupby(by='segment_key')['segment_osrm_distance'].cumsum()
df['segment_osrm_time_sum'] = df.groupby(by='segment_key')['segment_osrm_time'].cumsum()
```
